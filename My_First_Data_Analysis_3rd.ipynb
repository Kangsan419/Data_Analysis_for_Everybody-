## 인구 구조를 다양한 형태로 시각화하기

```python
#원하는 동네 인구 데이터 시각화 
import csv
import matplotlib.pyplot as plt

f=open('age.csv', encoding='949')
data = csv.reader(f)
result=[]
a = input('인구 구조가 알고 싶은 이름을 입력해라 : ')


for row in data :
    if a in row[0]:
        for i in row[3:]:
            result.append(int(i))
            
plt.style.use('seaborn-bright')
plt.rc('font', family ='AppleGothic') #plt.rc
plt.title(a + ' 지역 인구 구조') #plt.title
plt.plot(result)
plt.show()
```

### bar() 함수

함수에 입력되는 값

    ex)

    높이  1, 2, 3, 5, 6, 7

## 항아리 모양 그래프 그리기

```python
#성별에 따른 인구 분포 데이터 다루기 
import csv

f = open('gender.csv', encoding='949')
data = csv.reader(f)
m=[]
f=[]

for row in data:
    if '신도림' in row[0]:
        for i in range(0,101):
            m.append(int(row[i+3])) #3~104번까지 남자 나이 
            f.append(int(row[-(i+1)])) #-1 ~ -102번 
f.reverse()
```

```python
#데이터 시각화하기 
import csv

f = open('gender.csv', encoding='949')
data = csv.reader(f)
m=[]
f=[]

for row in data:
    if '신도림' in row[0]:
        for i in range(0,101):
            m.append(int(row[i+3])) #3~104번까지 남자 나이 
            f.append(int(row[-(i+1)])) #-1 ~ -102번 
f.reverse()

import matplotlib.pyplot as plt

plt.barh(range(101),m)
plt.barh(range(101),f)
plt.show()
```

```python
#항아리 모양으로 만들기 
import csv

f = open('gender.csv', encoding='949')
data = csv.reader(f)
m=[]
f=[]

for row in data:
    if '신도림' in row[0]:
        for i in range(0,101):
            m.append(-int(row[i+3])) #3~104번까지 남자 나이 
            f.append(int(row[-(i+1)])) #-1 ~ -102번 
f.reverse()

import matplotlib.pyplot as plt

plt.rcParams['axes.unicode_minus'] = False #마이너스 부호 깨짐 
plt.rc('font',family ='AppleGothic')
plt.title('신도림 지역의 남녀 성별 인구 분포')
plt.barh(range(101),m, label = '남성')
plt.barh(range(101),f, label ='여성')
plt.legend()
plt.show()
```

```python
#원하는 지역 성별 및 나이 확인하기 
import csv

f = open('gender.csv', encoding='949')
data = csv.reader(f)
m=[]
f=[]
neigh = input('동네 이름 입력 : ')

for row in data:
    if neigh in row[0]:
        for i in range(0,101):
            m.append(-int(row[i+3])) #3~104번까지 남자 나이 
            f.append(int(row[-(i+1)])) #-1 ~ -102번 
f.reverse()

import matplotlib.pyplot as plt

plt.rcParams['axes.unicode_minus'] = False #마이너스 부호 깨짐 
plt.rc('font',family ='AppleGothic')
plt.title(neigh + ' 지역의 남녀 성별 인구 분포')
plt.barh(range(101),m, label = '남성')
plt.barh(range(101),f, label ='여성')
plt.legend()
plt.show()
```

## 우리 동네 인구 구조를 파이 차트로 나타내기 

```python
#제주도에는 여성의 비율리 더 높을까 

import csv
f = open('gender.csv', encoding='949')
data = csv.reader(f)


m = []
f = []



name = input('찾고 싶은 지역의 이름을 알려주세요 :' )
for row in data :
    if name in row[0] :
        for i in row[3:104] :
            m.append(-int(i))
        for i in row[106:] :
            f.append(int(i))
        break #첫 번째로 만나는 데이터만 추가 시킬 수 있도록 함


import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name+ '지역의 남녀 성별 인구 분포')

#'제주특별자치도'를 검색하면 예하 모든 동네가 list에 append되어 error를 일으킴 
plt.barh(range(101), m, label='남성') 
plt.barh(range(101), f, label='여성')
plt.legend()
plt.show()
```

```python
#제주도 성별 파이차트로 구하기

import csv
import matplotlib.pyplot as plt 

f = open('gender.csv', encoding = '949')
data = csv.reader(f)
size=[]
name = input('확인하고 싶은 지역 이름 :')

for row in data :
    if name in row[0]:
        m = 0
        f = 0
        for i in range(101):
            m += int(row[i+3])
            f += int(row[i+106])
        break
size.append(m)
size.append(f)

plt.rc('font', family='AppleGothic')
color=['crimson','darkcyan']
plt.axis('equal')
#startangle => 시작 각도 조정, 90 -> 반시계방향으로 90도 
plt.pie(size, labels =['남','여'], autopct = '%.1f%%',colors = color, startangle =90)
plt.title(name + ' 지역 남녀 성비')
plt.show()
```

## 우리 동네 인구 구조를 산점도로 나타내기

### 꺾은선 그래프로 표현하기

```python
#제주도 나이별 성별 확인하기 
import csv
import matplotlib.pyplot as plt 

f =open('gender.csv', encoding='949')
data = csv.reader(f)
m =[]
f=[]

neigh = input('궁금한 지역 이름 : ')
for row in data :
    if neigh in row[0]:
        for i in range(3,104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
        break
        
plt.plot(m, label='male')
plt.plot(f, label='female')
plt.legend()
plt.show()
```

```python
#남성 인구수에서 여성 인구수를 뺀 차이를 막대 그래프로 표현하기 
import csv
import matplotlib.pyplot as plt 

f =open('gender.csv', encoding='949')
data = csv.reader(f)
result=[]

neigh = input('궁금한 지역 이름 : ')
for row in data :
    if neigh in row[0]:
        for i in range(3,104):
            result.append(int(row[i])-int(row[i+103]))
        break
        
plt.bar(range(101),result)
plt.legend()
plt.show()
```

### 제주도의 연령대별 성별 비율을 산점도로 표현하기

```python
import csv
f =open('gender.csv', encoding='949')
data = csv.reader(f)
m = []
f = []
name = input('궁금한 동네를 입력해주세요 : ')
for row in data : 
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
        break
        
import matplotlib.pyplot as plt
plt.scatter(m,f,c=range(101),alpha=0.5,cmap='jet') #컬러맵 적용
plt.colorbar()
plt.plot(range(max(m)),range(max(m)),'g') #추세선 추가
plt.show()
```

```python
#제주도의 연령대별 성별 비율을 산점도로 표현하기 

import csv
import math
f = open('gender.csv', encoding = '949')
data = csv.reader(f)
m=[]
f=[]
size=[]
name=input('궁금한 동네를 입력해주세요 : ')

for row in data :
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
            size.append(math.sqrt(int(row[i])+int(row[i+103])))
        break
        
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font', family='AppleGothic')
plt.figure(figsize=(10,5),dpi =300)
plt.title(name + ' 지역의 성별 인구 그래프')
plt.scatter(m,f, s=size, c=range(101), alpha=0.5, cmap='jet')
plt.colorbar()
plt.plot(range(max(m)),range(max(m)),'g')
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
            
```
