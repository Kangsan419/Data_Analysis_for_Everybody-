```python
#어떤 조건에 부합하는 데이터만 선별하는 mask 기능 
import numpy as np
a = np.arange(-5,5)
print(a,'\n')

# 특정 조건에 부함하는 mask 생성 
print(a<0,'\n')

#mask를 다시 배열에 적용
print(a[a<0],'\n')

#mask를 변수에 넣기
mask=abs(a)>3 #absolute value 
print(a[mask],'\n')

#mask 연결해서 사용
mask1 = abs(a)>3
mask2 = abs(a)%2 ==0
print(a[mask1+mask2]) #둘 중 하나의 조건이라도 참일 경우 
print(a[mask1*mask2]) #두 가지 조건이 모두 참일 경우
```

4. 궁금한 지역의 인구 구조와 가장 비슷한 인구 구조를 가진 지역 찾기

```python
#전국의 모든 지역에 대해 반복하며 그 차이가 가장 작은 지역을 찾는다
import numpy as np
import csv

f = open('age.csv', encoding = '949')
data = csv.reader(f)
next(data)
data = list(data)
name = input('인구 구조가 알고 싶은 지역의 이름 입력 :')
mn = 1 
result_name = ''
result = 0

for row in data:
    for i in range(3,103):
        row[i]=int(row[i])


for row in data:
    if name in row[0]:
        home = np.array(row[3:])/row[2] #사용자에게 입력 받은 지역의 인구 비율 저장
        
for row in data: #전체 데이터 반복
    away = np.array(row[3:])/row[2] #home에 저장된 값과 away에 저장된 값의 차이 함 계산 
    s = np.sum(home-away) #계산한 값이 최솟값인지 확인 
    if s<mn:
        mn = s #최솟값일 경우 mn의 값 갱신 
        result_name = row[0]
        result = away
        
#시각화 
import matplotlib.pyplot as plt
plt.rc('font', family='AppleGothic')
plt.title(name + '지역과 가장 비슷한 인구 구조를 가진 지역')
plt.plot(home, label=name)
plt.plot(result, label = result_name)
plt.legend()
plt.show()
```

```python
#우리 동네와 인구 구조가 가장 비슷한 동네를 찾는 코드
import numpy as np
import csv

#데이터 읽어오기 
f = open('age.csv', encoding = '949')
data = csv.reader(f)
next(data)
data= list(data)

#궁금한 지역의 이름을 입력 받는다
name=input('인구 구조가 궁금한 지역의 이름(읍면동 단위)을 입력 :')
mn = 1
result_name = ''
result = 0

#궁금한 지역의 인구 구조를 저장
for row in data:
    if name in row[0]:
           home = np.array(row[3:],dtype=int)/int(row[2])

#궁금한 지역의 인구 구조와 가장 비슷한 인구 구조를 가진 지역 찾기
for row in data:
    away = np.array(row[3:],dtype=int)/int(row[2])
    s = np.sum((home-away)**2)
    if s<mn and name not in row[0]:
        mn = s
        result_name = row[0]
        result = away
#시각화 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(10,5),dpi=300)
plt.rc('font',family='AppleGothic')
plt.title(name + '지역과 가장 비슷한 인구 구조를 가진 지역')
plt.plot(home, label=name)
plt.plot(result,label=result_name)
plt.legend()
plt.show()
```
